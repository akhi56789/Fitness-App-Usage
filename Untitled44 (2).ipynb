```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
data= pd.read_csv("/content/fitness_app_usage_dataset.csv")
print(data.head())
```

```python
#Basic info
print(data.info())
```

```python
#Cheek for missing value
print(data.isnull().sum())
```

```python
#Summary statistics
print(data.describe())
```

```python
#Unique values in categorical colomns
print("Unique Catagories:",data['Category'].unique())
print("Unique Labels:", data['Label'].unique())
```

```python
# drop rows with missinf values
data = data.dropna()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
# cheek for duplicate rows
data = data.drop_duplicates()
```

```python
# validate data type
print(data.dtypes)
```

```python
total_days=data.groupby('Category')['Active Days'].sum()
print(total_days)
```

```python
average_days = data.groupby('Label')['Active Days'].mean()
print(total_days)
```

```python
max_days=data.groupby('Label')['Active Days'].max()
print(max_days)
```

```python

data['Month'] = data['Date'].dt.month
monthly_trends = data.groupby('Month')['Active Days'].sum()
print(monthly_trends)
```

```python
# Visualaization
plt.figure(figsize=(8, 6))
plt.pie(total_days.values, labels=total_days.index, autopct='%1.1f%%', startangle=90)
plt.title("Total Active days by Category",  fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(average_days.index, average_days.values,color='green')
plt.title('average days by Category' ,fontsize=14)
plt.xlabel("Cetagory", fontsize=12)
plt.ylabel('Average_days', fontsize= 12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(monthly_trends.index, monthly_trends.values, marker='o', color='blue')
plt.title("Monthly Trends by Category", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Days", fontsize=12)
plt.show()
```
